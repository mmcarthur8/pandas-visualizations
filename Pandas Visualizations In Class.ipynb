```python
# import packages to create visualizations 
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
# Define Start/End dates, and ticker names
start1 = datetime.datetime(2022, 1, 1)
end1 = datetime.datetime(2022, 2, 15)
# Amazon
AMZN1 = data.DataReader("AMZN", 'yahoo', start1, end1)
```

```python
AMZN1
```

```python
myAMZN1 = AMZN1.reset_index()
```

```python
myAMZN1
```

```python
# use ggplot style
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
plt.style.use('ggplot')
myAMZN1.plot.line(x='Date',y='Volume', xlabel='Date', ylabel='AMZN Volume', figsize=(12,3),lw=1)
```

```python
# for month of december
start2 = datetime.datetime(2021, 12, 1)
end2 = datetime.datetime(2021, 12, 31)
# Amazon
AMZN2 = data.DataReader("AMZN", 'yahoo', start2, end2)
```

```python
myAMZN2 = AMZN2.reset_index()
myAMZN2.plot.bar(x='Date', y='Open')
```

```python
# Microsoft
# use corresponding data frame with january 1st to current date
MSFT1 = data.DataReader("MSFT", 'yahoo', start1, end1)
myMSFT1 = MSFT1.reset_index()
```

```python
myMSFT1
```

```python
# create time series line plot for january 1, 2022 to current for MSFT volume
plt.style.use('bmh')
myMSFT1.plot.line(x='Date',y='Volume', xlabel='Date', ylabel='MSFT Volume', figsize=(12,3),lw=1)
```

```python
# change start and end times for month of january 2022
start3 = datetime.datetime(2022, 1, 1)
end3 = datetime.datetime(2022, 1, 31)
# Microsoft
MSFT2 = data.DataReader("MSFT", 'yahoo', start3, end3)
# reset index to make sure date isn't the first index column
myMSFT2 = MSFT2.reset_index()
```

```python
# plot high and low densities for month of january for microsoft
myMSFT2[['High','Low']].plot.density()
```
